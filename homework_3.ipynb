```python
import numpy as np
import matplotlib.pyplot as plt
X = np.loadtxt("http://scipy-lectures.org/_downloads/populations.txt")
y = X[:,0] #years
h = X[:,1] #hares
l = X[:,2] #lynxes
c = X[:,3] #carrots
```

```python
plt.xlabel('Year')
plt.ylabel('Population')
plt.plot(y, h, label = "hares", color='r')
plt.plot(y, l, label = "lynxes", color='g')
plt.plot(y, c, label = "carrots", color='b')
plt.legend()
plt.show()
```

```python
s = h+l 
H = 100 * h/s
L = 100* l/s
plt.bar(y, H, label = "Hare")
plt.bar(y,L,  bottom = H,label = "Lynxes")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3) 
ax[0].hist(h, color = 'g', label='hare', bins=10)
ax[0].axvline(h.mean(), c="black")
plt.legend()
ax[1].hist(l, color = 'r', label='lynx')
ax[1].axvline(l.mean(), c="black")
plt.legend()
ax[2].hist(c, color = 'b', label='carrot')
ax[2].axvline(c.mean(), c="black")
plt.legend()
plt.show()
```

```python
plt.plot(y, h, label = "hare", c = 'g')
plt.plot(y, l, label = "lynx", c = 'r')
plt.plot(y, c, label = "carrot", c ='b')
plt.scatter(y[np.argmax(h)],np.max(h), c = 'g')
plt.scatter(y[np.argmax(l)],np.max(l), c = 'r')
plt.scatter(y[np.argmax(c)],np.max(c), c ='b')
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
a = np.where(h - l > 0, 'H', 'L')
p = np.array([len(np.argwhere(a == 'H')), len(np.argwhere(a == 'L'))])
plt.pie(p, labels=['hare', 'lynx'], autopct='%1.f%%')
plt.show()
print(a)
```

```python
plt.xlabel('Year')
plt.ylabel('Population')
plt.plot(y, h, label = "hare", c = 'g')
plt.plot(y, l, label = "lynx", c = 'r')
plt.plot(y, c, label = "carrot", c ='b')
plt.axhline(y=50000, c='black')
plt.scatter(y[h > 50000], h [h > 50000], c = 'g')
plt.scatter(y[l > 50000], l [l > 50000], c = 'r')
plt.scatter(y[c > 50000], c [c > 50000], c='b')
plt.legend()
plt.show()
```

```python
print("lowest pop of hares : ", y[h.argsort()[:2]])
print("lowest pop of lynxes : ", y[l.argsort()[:2]])
print("lowest pop of carrots : ", y[c.argsort()[:2]])
```

```python
plt.xlabel('Year')
plt.ylabel('Number')

plt.plot(y, np.gradient(h), label='Hare gradient', color='r')
plt.plot(y, l, label = "Lynxes", color='g')
plt.legend()
plt.show()
print("correlation coefficient =", np.corrcoef(np.gradient(h), l)[0, 1])
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 10))
ax[0].bar(y - 0.2, h, 0.4, color='g', label='Hare')
ax[0].bar(y + 0.2, l, 0.4, color='orange', label='Lynxes')
ax[0].legend()
ax[1].bar(y - 0.2, h, 0.4, color='g', label='Hare')
ax[1].bar(y + 0.2, c, 0.4, color='orange', label='Carrots')
ax[1].legend()
plt.show()
```
